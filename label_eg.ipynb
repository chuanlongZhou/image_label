```python
import os
import matplotlib.pyplot as plt
from constant import LABELS
```

```python
labeled = []
with open('data\Chandrapur\ground_labeled.txt', 'r') as f:
    for line in f:
        labeled.extend(line.strip().split(','))
```

```python
data_path = 'data\Chandrapur'
```

```python
def plot_mask(data_path, file):
    img = plt.imread(os.path.join(data_path, 'images', file))
    masks = []
    for label in LABELS:
        masks.append(plt.imread(os.path.join(data_path, 'masks', label, file)))
                     
    plt.imshow(img)
    colors = ['Oranges_r', 'gray', 'Blues_r', 'Reds_r']
    # add the mask with only value==1
    for index, mask in enumerate(masks):
        # replace 0 with nan
        mask[mask == 0] = None
        plt.imshow(mask, cmap=colors[index], alpha=0.5)
        
    plt.axis('off')
    plt.tight_layout()
    plt.title(file)
    plt.show()
```

```python
for file in labeled:
    plot_mask(data_path, file)
```
